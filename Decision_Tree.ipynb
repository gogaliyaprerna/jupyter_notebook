**Hue**



**What Does hue Do?**
The hue parameter allows you to assign different colors to data points based on the levels of a categorical variable.
It visually separates groups within your data, making it easier to identify patterns or differences.

**How hue Works in Seaborn**
1. In Univariate Plots (e.g., histograms, box plots):
Colors are assigned to different levels of the categorical variable.
2. In Bivariate/Multivariate Plots (e.g., scatter plots, pair plots):
Each group is plotted with a unique color.
